```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('gccmbf2013-exchange-rate-dolla-rial-november-2011-june-2020-en.csv')
```

```python
data.head(10)
```

```python
data.tail(10)
```

```python
plt.plot(data['persian calander dates'], data['last price recorded on the day'])
plt.title('Normal Form')
plt.rcParams["figure.figsize"] = (50,3)
plt.show() 
```

```python
plt.loglog(data['persian calander dates'], data['last price recorded on the day'])
plt.title('Log Form', fontdict={'fontsize': 50}) 
plt.rcParams["figure.figsize"] = (50, 10)
plt.show() 
```
